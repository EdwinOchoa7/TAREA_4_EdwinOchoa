```python
# Paso 3: Preprocesamiento
# Elimino columnas irrelevantes (por ejemplo, "CustomerID" si no aporta al análisis)
datos = datos.drop(columns=['CustomerID'], errors='ignore')
```

```python
# Transformo variables categóricas a numéricas (por ejemplo, "Genre")
datos['Gender'] = datos['Gender'].map({'Male': 0, 'Female': 1})
```

```python
# Escalamos los datos para que todas las variables tengan la misma escala
scaler = StandardScaler()
datos_escalados = scaler.fit_transform(datos)
```

<span style="font-size:50px; color:blue;">K-means<br>

```python
# Paso 4: K-means Clustering
# Pruebo diferentes valores de K y calculo la inercia
inercias = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(datos_escalados)
    inercias.append(kmeans.inertia_)
```

```python
# Grafico la "Elbow Method" para encontrar el valor óptimo de K
plt.figure(figsize=(8, 5))
plt.plot(k_values, inercias, marker='o', linestyle='--', color='purple')
plt.title('Método del Codo para Determinar K')
plt.xlabel('Número de Clústeres (K)')
plt.ylabel('Inercia')
plt.show()
```

```python
# Entreno el modelo con el K óptimo (por ejemplo, K=5)
kmeans_optimo = KMeans(n_clusters=5, random_state=42)
kmeans_optimo.fit(datos_escalados)
labels_kmeans = kmeans_optimo.labels_
```

```python
# Evaluo el modelo con métricas
silhouette_kmeans = silhouette_score(datos_escalados, labels_kmeans)
calinski_kmeans = calinski_harabasz_score(datos_escalados, labels_kmeans)
print(f"Coeficiente de Silhouette (K-means): {silhouette_kmeans:.2f}")
print(f"Índice de Calinski-Harabasz (K-means): {calinski_kmeans:.2f}")
```

```python
# Visualizo los clusters
plt.figure(figsize=(8, 5))
plt.scatter(datos_escalados[:, 0], datos_escalados[:, 1], c=labels_kmeans, cmap='viridis')
plt.title('Clusters de K-means')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.colorbar()
plt.show()
```

<span style="font-size:50px; color:blue;">Clustering Jerárquico<br>

```python
# Paso 5: Clustering Jerárquico
# Uso el método de linkage para calcular las distancias entre puntos
linkage_matrix = linkage(datos_escalados, method='ward')
```

```python
# Grafico el dendrograma
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix)
plt.title('Dendrograma - Clustering Jerárquico')
plt.xlabel('Clientes')
plt.ylabel('Distancia')
plt.show()
```

```python
# Corto el dendrograma para obtener clusters
num_clusters = 5
labels_hierarchical = fcluster(linkage_matrix, num_clusters, criterion='maxclust')
```

```python
# Evaluo el modelo jerárquico con métricas
silhouette_hierarchical = silhouette_score(datos_escalados, labels_hierarchical)
calinski_hierarchical = calinski_harabasz_score(datos_escalados, labels_hierarchical)
print(f"Coeficiente de Silhouette (Clustering Jerárquico): {silhouette_hierarchical:.2f}")
print(f"Índice de Calinski-Harabasz (Clustering Jerárquico): {calinski_hierarchical:.2f}")
```

```python
# Visualizo los clusters jerárquicos
plt.figure(figsize=(8, 5))
plt.scatter(datos_escalados[:, 0], datos_escalados[:, 1], c=labels_hierarchical, cmap='plasma')
plt.title('Clusters - Clustering Jerárquico')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.colorbar()
plt.show()
```
